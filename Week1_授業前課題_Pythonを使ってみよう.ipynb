```
"""
紙を43回折った時の厚さを計算するコード（for文を使って結果をリストに保存）
"""
THICKNESS = 0.00008
base = 2
exponent = 43
result_history = []
result_history.append(THICKNESS)
for i in range(exponent):
  folded_thickness = THICKNESS*(base**(i+1))
  result_history.append(folded_thickness)
print(len(result_history))
```

```
"""
紙を43回折った時の厚さを計算するコード（matplotlibを使ってグラフにする。）
"""
THICKNESS = 0.00008
base = 2
exponent = 43
result_history = []
result_history.append(THICKNESS)
for i in range(exponent):
  folded_thickness = THICKNESS*(base**(i+1))
  result_history.append(folded_thickness)

"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(result_history) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

紙を折った回数が増えれば増えるほど厚さの増加量が大きくなっている。
４０回折った段階では１０万km程度であったが、最後のわずか数回で６０万kmも増加し、月まで達している。

# Week2 授業前課題2 ダミーデータの作成  
機械学習に関するプログラムの実験を行う上で、実データではないランダムなデータ（ダミーデータ）が欲しいということがあります。正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていきましょう。

## 【問題1】乱数の作成  
>平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。  
>\[\left[ \begin{array}{ccc} 1.0 & 0.8\\ 0.8 & 1.0 \end{array} \right]\]  
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。  
>作成したndarrayのshapeは(500, 2)になります。

```python
import numpy as np

# 平均(乱数作成用)
mean_0 = [-3, 0]
# 共分散行列(乱数作成用)
cov_0 = [[1.0, 0.8], [0.8, 1.0]] 

# シードを ０ に設定して発生する乱数が実行ごとに変わらないようにする
np.random.seed(0)

# 平均が(-3, 0)、共分散行列[[1.0, 0.8], [0.8, 1.0]]の乱数５００個の行列を作成
random_ndarray_0 = np.random.multivariate_normal(mean_0, cov_0, 500)

print("作成された乱数行列：\n{}".format(random_ndarray_0))
print("作成された乱数行列の形は：{}".format(random_ndarray_0.shape))
```

## 【問題2】散布図による可視化  
>問題1で作成したデータ点を散布図により可視化してください。  
>散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt

# 散布図で可視化する
%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(random_ndarray_0[:,0], random_ndarray_0[:,1])
plt.show()
```

## 【問題3】ヒストグラムによる可視化  
>問題1で作成したデータをヒストグラムにより可視化してください。  
>ヒストグラムはplt.hist()を使うことで描けます。  
>各次元ごとに作成し、2つのヒストグラムとしてください。  
>それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
import matplotlib.pyplot as plt

# ヒストグラムで可視化する
%matplotlib inline

# x1をヒストグラム化x軸：-6 ~ 3、棒の数：50）
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")
plt.xlim(right=3, left=-6)
plt.hist(random_ndarray_0[:,0], bins=50)
plt.show()

# x2をヒストグラム化（x軸：-6 ~ 3、棒の数：50）
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.xlim(right=3, left=-6)
plt.hist(random_ndarray_0[:,1], bins=50)
plt.show()
```

## 【問題4】データの追加  
>新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。  
>\[\left[ \begin{array}{ccc} 1.0 & 0.8\\ 0.8 & 1.0 \end{array} \right]\]    
>そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
import numpy as np
import matplotlib.pyplot as plt


# 平均(新しい乱数作成用)
mean_1 = [0, -3]
# 共分散行列(新しい乱数作成用)
cov_1 = [[1.0, 0.8], [0.8, 1.0]] 

# シードを ０ に設定して発生する乱数が実行ごとに変わらないようにする
# np.random.seed(0)

# 平均が(0, -3)、共分散行列[[1.0, 0.8], [0.8, 1.0]]の乱数５００個の行列を作成
random_ndarray_1 = np.random.multivariate_normal(mean_1, cov_1, 500)


# 散布図で可視化する
%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

plt.scatter(random_ndarray_0[:,0], random_ndarray_0[:,1], label='0')
plt.scatter(random_ndarray_1[:,0], random_ndarray_1[:,1], label='1')
plt.legend(loc='upper right')
plt.show()
```

## 【問題5】データの結合  
>データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
import numpy as np

# 乱数のndarrayを結合する(concatenate)
concatenate_ndarray = np.concatenate((random_ndarray_0, 
                                      random_ndarray_1), axis=0)

# 乱数のndarrayを結合する(.vstack 縦方向に結合)
vstack_ndarray = np.vstack((random_ndarray_0, random_ndarray_1))


# 答え
print("concatenateで作成されたndarrayの形は：{}".format(concatenate_ndarray.shape))
print("vstackで作成されたndarrayの形は：{}".format(vstack_ndarray.shape))
```

## 【問題6】ラベル付け  
>ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。  
>0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。  
>機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
import numpy as np

# random_ndarray_0 と同じサイズのラベル用ndarray（０のみ）を作成
zeros_for_ndarray_0 = np.zeros((len(random_ndarray_0[:,0]), 1))
# random_ndarray_1 と同じサイズのラベル用ndarray（１のみ）を作成
ones_for_ndarray_1 = np.ones((len(random_ndarray_1[:,0]), 1))

# hstackを使ってラベル用ndarrayを結合する
labeled_ndarray_0 = np.hstack((random_ndarray_0, zeros_for_ndarray_0))
labeled_ndarray_1 = np.hstack((random_ndarray_1, ones_for_ndarray_1))

# 乱数のndarrayを結合する(concatenate)
concatenate_ndarray = np.concatenate((labeled_ndarray_0, 
                                      labeled_ndarray_1), axis=0)

# 乱数のndarrayを結合する(.vstack 縦方向に結合)
vstack_ndarray = np.vstack((labeled_ndarray_0, labeled_ndarray_1))

# 答え
print("concatenateで作成されたラベル付きのndarrayの形は：{}"
      .format(concatenate_ndarray.shape))
print("vstackで作成されたラベル付きのndarrayの形は：{}"
      .format(vstack_ndarray.shape))
```
